```python
def parse_log_entry(log_entry):
    ip_pattern = r'\d+\.\d+\.\d+\.\d+'
    ip_match = re.search(ip_pattern, log_entry)
    if ip_match:
        return ip_match.group()
    else:
        return None
    
```
